```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
observed = np.random.randint( 15, size=5)

print(f'Observed values: {observed}')

total = np.sum(observed)
expected = np.full_like(observed, total / len(observed))
chi_square_stat = np.sum(((observed - expected) ** 2) / expected)
df = len(observed) - 1
p_value = 1 - stats.chi2.cdf(chi_square_stat, df)
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

print(f'Chi-square statistic: {chi_square_stat:.2f}')
print(f'Degrees of freedom: {df}')
print(f'Critical value (0.05 significance level): {critical_value:.2f}')
print(f'P-value: {p_value:.4f}')
```

```python
plt.bar(range(1, len(observed) + 1), observed, width=0.5, label='Observed', alpha=0.8 , color ='black' )
plt.bar(np.arange(1, len(observed) + 1) + 0.5, expected, width=0.5, label='Expected', alpha=0.8 , color ='green')
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.show()
```
